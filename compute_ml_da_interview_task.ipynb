```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("breast_cancer_survival.csv")
```

```python
df.head(31)
```

```python
df.isnull()
```

```python
null_percentages = (df.isnull().sum() / len(df)) * 100
for column in df.columns:
    if null_percentages[column] > 10:
        df = df.drop(column, axis=1)
    else:
        df = df.dropna(subset=[column])
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

Visualizations

```python
tumour_stage_counts = df['Tumour_Stage'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(tumour_stage_counts, labels=tumour_stage_counts, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Tumour Stages')
plt.axis('equal') 
```

```python
df.value_counts("Tumour_Stage")
```

```python
tumor_stage_counts = df['Tumour_Stage'].value_counts()
plt.figure(figsize=(4, 4))
for stage in tumor_stage_counts.index:
    filtered_df = df[df['Tumour_Stage'] == stage]
    plt.hist(filtered_df['Tumour_Stage'], bins=len(tumor_stage_counts),
             alpha=0.8, edgecolor='black', label=stage)
plt.xlabel('Tumor Stage')
plt.ylabel('Count')
plt.title('Histogram of Tumor Stages with Color Hue')
# Show the legend
plt.legend(title='Tumor Stage')
# Show the histogram
plt.show()
```

```python
df.rename(columns={'Patient_Status': 'Target'}, inplace=True)
df['Target'] = df['Target'].map({'Alive': 1, 'Dead': 0})
df
```

```python
age_bins = [0, 20, 40, 60, 80, 100]
df['age_bins'] = pd.cut(df['Age'], bins=age_bins, labels=['0-20', '21-40', '41-60', '61-80', '81-100'])
df
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age_bins'], bins=len(age_bins), color='skyblue', alpha=0.8 )
plt.xlabel('Age Bins')
plt.ylabel('Count')
plt.title('Histogram of Count of People in Each Age Bin');
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Male and Female Patients')
plt.xticks(rotation=0);
```

The data has more values of one gender over the other, here are some possible reasons:

1. **Breast Cancer Prevalence:** Breast cancer is more prevalent in females compared to males. This natural difference in prevalence could lead to an unequal distribution of gender in a dataset related to breast cancer.

2. **Medical Presentation:** Breast cancer might present differently in males and females. This could lead to differences in diagnosis rates and hence, a larger representation of one gender over the other in a dataset.

3. **Healthcare Seeking Behavior:** Societal norms and expectations might influence healthcare-seeking behavior. Females may be more likely to undergo breast cancer screenings and seek medical attention, leading to earlier diagnoses and more data for females.

4. **Data Collection Bias:** There could be biases in data collection methods. For example, if data was collected primarily from hospitals or clinics that predominantly serve females, it could result in an overrepresentation of female patients.

5. **Clinical Trials and Research Focus:** Historically, breast cancer research and clinical trials have focused on female patients due to their higher prevalence. This focus could lead to more available data for females compared to males.

6. **Underreporting in Males:** Males with breast cancer might face stigma and social taboos that discourage reporting of the condition. This could result in underreporting and fewer male patients in the dataset.

7. **Data Collection Period:** The time period during which the data was collected might also play a role. Awareness campaigns and increased medical attention might have led to increased diagnoses among females during that period.

8. **Biological Differences:** Gender differences at the biological level might influence the risk of developing breast cancer. Hormonal differences and genetic factors could contribute to varying susceptibility.

9. **Data Availability:** Availability of data might be influenced by factors such as access to healthcare, insurance coverage, and geographic location, which could impact the gender distribution.

In summary, the unequal distribution of genders in breast cancer data can arise from a combination of biological differences, medical practices, societal norms, biases in data collection, and the prevalence of the disease in the population. It's essential to consider these factors when interpreting and analyzing such data.

```python
column_to_convert = 'Histology'
df[column_to_convert] = df[column_to_convert].astype('string')
```

```python
df_encoded = pd.get_dummies(df, columns=["Histology"], prefix=["Histology"])
```

```python
columns_to_select = ['Age', 'Protein1', 'Protein2', 'Protein3', 'Protein4']
encoded_histology_columns = [col for col in df_encoded.columns if col.startswith("Histology_")]
df_encoded[encoded_histology_columns] = df_encoded[encoded_histology_columns].astype(float)
selected_data = df_encoded[columns_to_select + encoded_histology_columns]
correlation_matrix = selected_data.corr()
# Creating a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm', edgecolors='w', linewidths=0.5)
plt.colorbar(heatmap, label='Correlation')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Heatmap between Age, Protein Levels, and Encoded Histology')
plt.tight_layout()
plt.show()
```

1. **Gender Imbalance in Breast Cancer Data:**

6. **Biological and Clinical Considerations:**
